#### 课堂练习： 爬取天气后报网站2014-2018年上海天气数据（日期、天气情况、温度、风力风向），补充完整缺失数据后，按日期顺序整合为DataFrame。

1. 爬取2014-2018年上海天气数据

```python
import re
import urllib.request
def get_weather(url):
    response=urllib.request.urlopen(url,timeout=60)
    html=response.read()
    htmlstr=html.decode('gbk')
    a1=re.findall('title="(.+?)上海天气预报"',htmlstr)  #日期
    a2=re.findall('\r\n +([\u4e00-\u9fa5]+)\r\n +/([\u4e00-\u9fa5]+)</td>',htmlstr) #天气
    a3=re.findall('<td>\r\n +(-?\d+)℃\r\n +/\r\n +(-?\d+)℃\r\n +</td>',htmlstr)    #温度
    a4=re.findall('<td>\r\n +([\u4e00-\u9fa5]+风) (.+?)级\r\n +/([\u4e00-\u9fa5]+风) (.+?)级</td>',htmlstr) #风力
    return a1,a2,a3,a4

date=[]
weather=[]
temp=[]
wind=[]
for year in range(2014,2019):
    for i in range(12):
        j=i+1
        baseurl_start='http://www.tianqihoubao.com/lishi/shanghai'
        if j<10:
            baseurl_end='/month/'+str(year)+'0'+str(j)+'.html'
        else:
            baseurl_end='/month/'+str(year)+str(j)+'.html' 
        url=baseurl_start+baseurl_end
        try:
            t1,t2,t3,t4=get_weather(url)
            date.append(t1)
            weather.append(t2)
            temp.append(t3)
            wind.append(t4)            
            print(str(year)+'年第'+str(j)+ '月抓取成功') 
        except:
            print(str(year)+'年第'+str(j)+ '月抓取失败') 
```

2.找出天气数据中有缺失值的月份

```python
#查询60个月中数据不一致的月份
for i in range(60):
    if (len(date[i]) !=len(weather[i])) | (len(date[i]) !=len(temp[i]))|(len(date[i]) !=len(wind[i])):
        print(str(i),len(date[i]),len(weather[i]),len(temp[i]),len(wind[i]))
```

3.将天气数据嵌套列表转换为非嵌套列表，补充缺失值

```python
#将嵌套的天气数据列表，转换为非嵌套列表
date1=[]
for i in date:
    date1=date1+i
weather1=[]
for i in weather:
    weather1=weather1+i
temp1=[]
for i in temp:
    temp1=temp1+i
wind1=[]
for i in wind:
    wind1=wind1+i  

#查找有缺失值这一天，在date1列表中对应的index序号
i=date1.index('2018年05月23日')
#向weather1, temp1,wind1这3个列表中序号i的位置手动添加缺失的数据
weather1.insert(i,('小雨','小雨'))
temp1.insert(i,(29,29))
wind1.insert(i,('北风', '4～5', '北风', '4～5'))
```

4.生成dataframe

```python
#整合等长的date1,weather1,temp1,wind1形成dataframe
import pandas as pd
dic={'date':date1,'weather':weather1,'temper':temp1,'wind':wind1}
df=pd.DataFrame(dic)
df
```
